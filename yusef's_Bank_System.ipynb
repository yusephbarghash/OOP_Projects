# Task 1 : 
- #### task 1.1 Make Class(User) with attributes:
    - name 
    - age 
    - gender 
    - balance
- #### task 1.2 Make method to show previous data 

```python
class User:
    listOfNames = []
    def __init__(self, n='n' , a=0 , g='a' , b=0):
        self.__name = n
        self.__age = a
        self.__gender = g
        self.__balance = b
        self.listOfNames.append(self.__name)
    def get_user_data(self):
        print(f"\n\n{self.__name} data is : ")
        print(f"1. {self.__name}'s age is {self.__age} ")
        print(f"2. {self.__name}'s Gender is {self.__gender}")
        print(f"3. {self.__name}'s balance is {self.__balance}")
```

# Task 2 : 
- Make Child Class(Bank) which inherite from Class(user) with Mehods:
    - #### task 2.1 deposite : which take parameter(amount) and add it to balance and print new balance.  
    - #### task 2.2 withdraw : which take parameter(amount) and check if user have enough money. 
    - #### task 2.3 view balance : to show current balance. 


```python
class Bank(User):
    def deposit(self , amount):
        #self.amount = amount
        self._User__balance =  self._User__balance + amount
        print(f"\nThe amount added successufully \nNow , the new {self._User__name}'s balance is {self._User__balance}")
    def withdraw(self, amount):
        if self._User__balance >= amount:
            self._User__balance = self._User__balance - amount
            print(f"\nThe amount withdrawed successufully \nNow , the new {self._User__name}'s balance is {self._User__balance}")
        else :
            print(f"You can't withdraw {amount}")
            print(f"Your balance is only {self._User__balance} , so please withdraw an appropiate amount")
            
    def Show_balance(self):
        print(f"Your balance is  {self._User__balance} ")
```
